```python
fraud_data['Output'] = np.where(fraud_data['Taxable.Income']<=30000,'Risky','Good')
fraud_data.head(50)
```

```python
#Creating dummy variable for the columns
fraud_data1=pd.get_dummies(fraud_data,columns=['Undergrad','Marital.Status','Urban'])
label_encoder = preprocessing.LabelEncoder()
fraud_data1['Output']= label_encoder.fit_transform(fraud_data1['Output']) 
```

```python
fraud_data1.head()
```

```python
column_names = ["Taxable.Income","City.Population","Work.Experience", "Undergrad_NO","Undergrad_YES","Marital.Status_Divorced","Marital.Status_Married","Marital.Status_Single","Urban_NO","Urban_YES","Output"]
```

```python
fraud_data2 = fraud_data1.reindex(columns=column_names)
```

```python
fraud_data2.head()
```

```python
fraud_data2=fraud_data2.drop('Taxable.Income',axis=1)
```

```python
fraud_data2.head()
```

```python
fraud_data2.shape
```

```python
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler
scaler = preprocessing.StandardScaler()
import warnings
warnings.filterwarnings("ignore")
sig_num_col=['City.Population','Work.Experience','Undergrad_NO','Undergrad_YES','Marital.Status_Divorced','Marital.Status_Married','Marital.Status_Single','Urban_NO','Urban_YES']
fraud_data2[sig_num_col] = scaler.fit_transform(fraud_data2[sig_num_col])
```

```python
X=fraud_data2.iloc[:,:9]
y=fraud_data2.iloc[:,9]
```

```python
X
```

```python
y
```

```python
fraud_data2['Output'].unique()
```

```python
# Splitting data into training and testing data set
X_train, X_test,y_train,y_test = train_test_split(X,y, test_size=0.2,random_state=40)
```

##  Using Random forest classifier

```python
from sklearn.ensemble import RandomForestClassifier as rfc
```

```python
rf_model = rfc(n_jobs=3,n_estimators=15,oob_score=True,criterion="entropy")
rf_model.fit(X_train,y_train)
```

## ------------------------------------------------------------------------------------------------------------------------

```python
# Training data
y_pred_train = rf_model.predict(X_train)
```

```python
y_pred_test = rf_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
#Training accuracy
accuracy_score(y_train, y_pred_train)
```

```python
confusion_matrix(y_train, y_pred_train)
```

```python
#Test accuracy
accuracy_score(y_test, y_pred_test)
```

```python
confusion_matrix(y_test, y_pred_test)
```

```python
print('Train accuracy of a model using dicision tree is : {0:0.4f}'. format(accuracy_score(y_train, y_pred_train)))
```

```python
print('Test accuracy of a model using dicision tree is : {0:0.4f}'. format(accuracy_score(y_test, y_pred_test)))
```
